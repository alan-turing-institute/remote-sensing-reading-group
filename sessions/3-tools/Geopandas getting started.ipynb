```python
import geodatasets
import geopandas as gpd
gdf = gpd.read_file(geodatasets.get_path('geoda.nyc_neighborhoods')) 
gdf['college_frac'] = gdf['college'] / gdf['poptot']
gdf.plot('college_frac', cmap='viridis', legend=True, 
         legend_kwds={'label': ">25 y/old with college degree (%)",})
```

```python
gdf = gpd.read_file(geodatasets.get_path('geoda.nyc_neighborhoods'))  # metadata: https://geodacenter.github.io/data-and-lab/NYC-Nhood-ACS-2008-12/
print(f'CRS: {gdf.crs}')
gdf.head(2)
```

```python
## Basic methods for geometry:
assert 'area' not in gdf.columns
area = gdf.area  # even though not in columns, it is a property of the geometry
plt.plot(area, gdf['poptot'], '.')
plt.xlabel('Area'); plt.ylabel('Population');
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 3))
for i_plot, col in enumerate(['popdty', 'comm_30_44', 'female']):
    gdf_plot = gdf[gdf[col] != 'NA']
    gdf_plot[col] = gdf_plot[col].astype(float)
    gdf_plot.plot(col, legend=True, ax=ax[i_plot])
```

```python
gdf_plot = gdf#[gdf['medianinco'] != 'NA']
gdf_plot['medianinco'] = gdf_plot['medianinco'].replace('NA', np.nan)
gdf_plot['medianinco'] = gdf_plot['medianinco'].astype(float)
gdf_plot.plot('medianinco', cmap='viridis', legend=True)
```

```python
ax = plt.subplot(111)
gdf.plot(facecolor='blue', edgecolor='none', ax=ax)
gdf.buffer(0.003).plot(edgecolor='black', facecolor='none', ax=ax)
```
